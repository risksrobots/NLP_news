Препроцессинг

```python
def preprocessing(data, lemmatizer = WordNetLemmatizer(), 
                  stop_words = stopwords.words('english')):
    texts = data.values.tolist()
    new_text = []

    for i in tqdm(range(len(texts))):
        text = texts[i]

        # приводим к нижнему регистру и дропаем стоп-слова
        punct_free = re.sub('[^a-zA-Z]', ' ', str(text).lower())
        punct_free = re.sub(r'\s.,_+', ' ', punct_free)
        stop_free = " ".join([i for i in punct_free.split() if i not in stop_words])
        lemma_text = ' '.join([lemmatizer.lemmatize(word) for word in stop_free.split(' ')])
        new_text.append(lemma_text)
    
    return new_text
```

```python
def transform_data(data, with_discret_col):
   
    text_title = preprocessing(data['title'])
    text_summary = preprocessing(data['summary'])
    data['title'] = text_title
    data['summary'] = text_summary
    
    tfidf_title = TfidfVectorizer(max_df=0.05, min_df=0.0009).fit_transform(data['title'])
    tfidf_summary = TfidfVectorizer(max_df=0.05, min_df=0.0009).fit_transform(data['summary'])
    df_title = pd.DataFrame(tfidf_title.todense())
    df_summary = pd.DataFrame(tfidf_summary.todense())
    
    if with_discret_col==1:
        merge_df = pd.concat([df_title, df_summary, data[['topic', 'subjectivity_score', 'target_shifted']]], axis=1)
    else:
        merge_df = pd.concat([df_title, df_summary], axis=1)
    return merge_df
```

# New ML

```python
metrics = {'F1_Score':f1_score, 'Accuracy': accuracy_score, 'Precision': precision_score, 'Recall': recall_score}

models_dict = {
    'LogisticRegression': LogisticRegression(), 
    'RandomForest': RandomForestClassifier(random_state=random_seed), 
    'DecisionTreeClassifier': DecisionTreeClassifier(random_state=random_seed),
    'GradientBoostingClassifier': GradientBoostingClassifier(random_state=random_seed)
}
```

```python
param_models = {
    'LogisticRegression': {'model__C': [0.0001, 0.001, 0.01, 0.1, 1]},
    'RandomForest': {'model__max_depth': [3, 5, 7], 'model__n_estimators':[50, 100, 150]},
    'DecisionTreeClassifier': {'model__min_samples_split': [3, 5, 7]} ,
    "GradientBoostingClassifier": {'model__learning_rate': [0.01]}
}

hyper_params = {}
```

```python
def calculate_metrics(y_test, y_pred, metrics={}):
    if len(metrics) == 0:
        raise ValueError('metrics are empty')
        
    results = dict()
    for key in metrics:
        results[key] = metrics[key](y_test, y_pred)
    
    return results

metrics = {'F1_Score':f1_score, 'Accuracy': accuracy_score, 'Precision': precision_score, 'Recall': recall_score}
```

```python
def training(data_train, target_train, data_test, target_test, name):
    result_df = pd.DataFrame([])
    for model_name, model in tqdm(models_dict.items()):
    
        params = param_models[model_name]

        experiment = Experiment(
            api_key="hEEHZKuye0VnYPCcJBDyazzeA",
            project_name="general",
            workspace="iis-diink",
        )
        experiment.set_name(f"{model_name}_{name}")

        pipeline = Pipeline([('model', model)])

        grid_search = GridSearchCV(pipeline, param_grid=params)
        grid_search.fit(data_train, target_train)

        best_model = grid_search.best_estimator_

        hyper_params[model_name] = grid_search.best_params_
        experiment.log_parameters(grid_search.best_params_)
        target_pred = best_model.predict(data_test)

        res_loc = calculate_metrics(target_test, target_pred, metrics=metrics)  

        experiment.log_metrics(res_loc)

        res_loc_df = pd.DataFrame(list(res_loc.values()), index=list(res_loc.keys()), columns=[model_name]).T
        result_df = pd.concat([result_df, res_loc_df])
        
    return hyper_params, result_df
```

```python
hyper_params_aug, metrics_df_aug = training(data_aug_train, target_aug_train, data_aug_test, target_aug_test, "aug_data")
hyper_params, metrics_df = training(data_train, target_train, data_test, target_test, "simple_data")
```
